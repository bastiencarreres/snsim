```python
import snsim
import numpy as np
import astropy.cosmology as acosmo
cosmo = acosmo.Planck18
import matplotlib.pyplot as plt
```

```python
def rate(z):
    return 1e-5

def _compute_zcdf(z_range):
    """Give the time rate SN/years in redshift shell.

    Parameters
    ----------
    z : numpy.ndarray
        The redshift bins.

    Returns
    -------
    numpy.ndarray(float)
        Numpy array containing the time rate in each redshift bin.

    """
    z_min, z_max = z_range

    # -- Set the precision to dz = 1e-5
    dz = 1e-5

    z_shell = np.linspace(z_min, z_max, int((z_max - z_min) / dz))
    z_shell_center = 0.5 * (z_shell[1:] + z_shell[:-1])
    co_dist = cosmo.comoving_distance(z_shell).value
    shell_vol = 4 * np.pi / 3 * (co_dist[1:]**3 - co_dist[:-1]**3)

    # -- Compute the sn time rate in each volume shell [( SN / year )(z)]
    shell_time_rate = rate(z_shell_center) * shell_vol / (1 + z_shell_center)

    z_pdf = lambda x : np.interp(x, z_shell, np.append(0, shell_time_rate))

    return snsim.utils.CustomRandom(z_pdf, z_min, z_max, dx=1e-5), (z_shell, shell_time_rate)


def _compute_zcdf2(z_range):
    """Give the time rate SN/years in redshift shell.

    Parameters
    ----------
    z : numpy.ndarray
        The redshift bins.

    Returns
    -------
    numpy.ndarray(float)
        Numpy array containing the time rate in each redshift bin.

    """
    z_min, z_max = z_range

    # -- Set the precision to dz = 1e-5
    dz = 1e-5

    z_shell = np.linspace(z_min, z_max, int((z_max - z_min) / dz))
    co_dist = cosmo.comoving_distance(z_shell).value
    shell_vol = 4 * np.pi / 3 * co_dist**2 * snsim.constants.C_LIGHT_KMS / cosmo.H(z_shell).value * dz

    # -- Compute the sn time rate in each volume shell [( SN / year )(z)]
    shell_time_rate = rate(z_shell) * shell_vol / (1 + z_shell)

    z_pdf = lambda x : np.interp(x, z_shell, shell_time_rate)

    return snsim.utils.CustomRandom(z_pdf, z_min, z_max, dx=1e-5), (z_shell, shell_time_rate)
```

```python
pdf, _ = _compute_zcdf((0.01, 0.1))
pdf2, _ = _compute_zcdf2((0.001, 0.1))

```

```python
plt.plot(pdf.x, pdf.pdfx)
plt.plot(pdf2.x, pdf2.pdfx)
```

```python
plt.plot(pdf.x, pdf.cdf)
plt.plot(pdf2.x, pdf2.cdf)
```

```python
plt.hist(pdf.draw(10000000), bins=500)
plt.hist(pdf2.draw(10000000), bins=500, histtype='step');
```
